```python
# Librería de análisis de datos
import pandas as pd
# Librería numérica especializada
import numpy as np
# Librería de gráficos generales
import matplotlib.pyplot as plt
# Librería de gráficos estadísticos
import seaborn as sns
```

```python
data = pd.read_excel("./Data.xls")
data
```

## Nivel 1: Operador
Operador 1, línea de producción 1

```python
data_operador = data[(data['Op'] == 1) & (data['LP'] == 1)]

data_operador
```

```python
ES = np.array([1.124]) # Especificación superior
EI = np.array([1.118]) # Especificación inferior
ED_operador = pd.DataFrame({
    
        "Media": data_operador[['D', 'SF']].mean(),
    
        "Mediana": data_operador[['D', 'SF']].median(),
    
        "Moda": data_operador[['D', 'SF']].mode().iloc[0,:],
    
        "Mitad del Rango": (data_operador[['D', 'SF']].max() + 
                            data_operador[['D', 'SF']].min())/2,
    
        "Desviación Estándar": data_operador[['D', 'SF']].std(),
    
        "Varianza": data_operador[['D', 'SF']].var(),
    
        "Mínimo": data_operador[['D', 'SF']].min(),
    
        "Máximo": data_operador[['D', 'SF']].max(),
    
        "Rango": data_operador[['D', 'SF']].max() - 
                 data_operador[['D', 'SF']].min(),
    
        "Coeficiente de Variación": (data_operador[['D', 'SF']].std() /
                                     data_operador[['D', 'SF']].mean() * 100),
    
        "Capacidad potencial": (ES-EI) / (6 * data_operador[['D']].std())                                        
    })
ED_operador
```

```python
# Se importa la librería de graficos estándar
import matplotlib.pyplot as plt
# se usa el estilo
plt.style.use('fivethirtyeight')
# se grafica en linea
%matplotlib inline

# Número de categorías para el histograma
numero_de_datos = data_operador.index.size
numero_de_categorias = np.int(np.rint(1 + 3.3 * np.log10(numero_de_datos)))

# Graficamos el histograma
data_operador[['D']].plot(kind='hist', bins=5)
```

```python
#plt.hold(True)
plt.plot(data_operador[['D']],marker='o',
         markerfacecolor='k',markersize=10,label='D')
plt.xticks([])
plt.ylabel('D [in]')

# Límites reales naturales
LRI = data_operador[['D']].mean() - 3 * data_operador[['D']].std()
LRS = data_operador[['D']].mean() + 3 * data_operador[['D']].std()
plt.plot([-1,10],[LRI,LRI],color='b',label='media - 3sd',alpha=0.50)
plt.plot([-1,10],[LRS,LRS],color='r',label='media + 3sd',alpha=0.50)

# Media
media = data_operador[['D']].mean()
plt.plot([-1,10],[media,media],color='g',label='Media',alpha=0.50)
plt.title("Gráfico de control")
plt.legend(loc=0,fontsize='small',markerscale=0.5)
```

```python
import seaborn as sns
sns.distplot(data_operador[['D']])
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-ticks')
plt.boxplot([data_operador[['D']]])
```

```python
# Área superficial
no_cumple = data_operador[['SF']].sum()
cumple = data_operador[['SF']].count() - data_operador[['SF']].sum()

# Gráfico circular
plt.figure(figsize=(5,5))
ax = plt.axes([0.1, 0.1, 0.8, 0.8])
explode = np.array([.2,0])
colors = sns.dark_palette("skyblue", 8, reverse=True)
labels = ['Cumple con AS', 'No cumple con AS']
ax.pie([cumple, no_cumple], explode = explode, 
       colors = colors, labels = labels, 
       autopct = '%2.0f%%', shadow = True)
```
